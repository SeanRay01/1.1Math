```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
def u(t):
    if t<0:
        return(0)
    else:
        return(1)
def e(t):
    return(np.exp(t))
def sin(t):
    return(np.sin(t))
def cos(t):
    return(np.cos(t))
def ten(t):
    return(np.tan(t))
```

```python
l=3/1000
ts=np.linspace(0.001,l,1000)
ft=list()
c=list()
l=list()
v=list()
i=list()
for t in ts:
    # ft.append((t**2)*(u(t-0)-u(t-2))+(1-t-4*t**2)*u(t-2))
    ft.append(2*sin(2*pi*1000*t))
    c.append(1/(2*pi*t*0.001))
    l.append((2*pi*t*0.001))
    v.append(20*e(-t/0.043))
    i.append(20/43000*e(-t/0.043))
    
```

```python
ft=list()
f=1000
t=2000
ts=np.linspace(1/t,5/t,1000)
for t in ts:
    # ft.append((t**2)*(u(t-0)-u(t-2))+(1-t-4*t**2)*u(t-2))
    ft.append(2*sin(2*pi*f*t))
plt.figure(figsize=(8,6))
plt.ylabel('v')
plt.xlabel('s')
plt.grid(True)
plt.plot(ts, ft, color='red')
plt.show()
```
