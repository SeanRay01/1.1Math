```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bag=[[],
     [],
     []]
for k in range(1,50+1):
    for j in range(1,15+1):
        for i in range(1,30+1):
                bag[0].append(i)
                bag[1].append(j)
                bag[2].append(k)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')	
ax.scatter3D(bag[0],bag[1],bag[2])
plt.show()
```
