```python
import matplotlib.pyplot as plt
import numpy as np
import math
def u(t):
    if t<0:
        return(0)
    else:
        return(1)
def extremum(x,y,title):
    ys=[0,[0],[10000]]
    xs=[0,[0],[10000]]
    for i,j in zip(y,x):
        if ys[1][0]<i:
            ys[1][0]=np.round(i,3)
            xs[1][0]=np.round(j,3)
        elif ys[2][0]>i:
            ys[2][0]=np.round(i,3)
            xs[2][0]=np.round(j,3)
    print('%s:x=%s,y=%s/x=%s,y=%s'%(title,xs[1][0],ys[1][0],xs[2][0],ys[2][0]))
```

```python
l=12#長度
x=np.linspace(0,l,1000)
v=list()
m=list()
for t in x:
    v.append((66-10*t)*(1-u(t-4)) + (36-10*t)*(u(t-4)-u(t-10)) +
             (120-10*t)*(u(t-10)-u(t-12)))  # 剪力方程式
    m.append((66*t-5*t**2)*(1-u(t-4)) + (120+36*t-5*t**2)*(u(t-4)-u(t-10)) +
             (-720+120*t-5*t**2)*(u(t-10)-u(t-12)))  # 彎矩方程式
extremum(x,v,'V(x)')
extremum(x,m,'M(x)')
```

```python
fig = plt.figure(figsize=(10,5))
fig.subplots_adjust(wspace=0.2, hspace=0.35)
for i in [1,2]:
    ax = fig.add_subplot(1,2, i )
    ax.set_xlim(0, l*1.1)
    ax.set_xlabel('x')
    ax.grid(True)
    if i == 1:
        ax.set_title('Shear force')
        ax.set_ylabel('V')
        ax.plot(x, v)
    elif i == 2:
        ax.set_title('Bending moment')
        ax.set_ylabel('M')
        ax.plot(x, m)
plt.show()
```
